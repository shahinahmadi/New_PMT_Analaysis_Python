```python
import math
import numpy
from matplotlib import pyplot as plt
import pandas as pd
import csv
from itertools import zip_longest

n_air=1
n_acr=1.4
n_water=1.33
n_glass=1.5

R_4=26
R_3=24
R_2=23.5
R_1=23

x=[]
Theta_1=[]
Theta_2=[]
l_1=[]
Phi_1=[]
d_1=[]
Theta_3=[]


new_data=pd.read_csv("Gain_position.csv")



for i in new_data["Radial position"]:
    x.append(i*100)
print("***************************************")
print("This is the position of the laser: ")
print(x)


#Calculation of Theta_1
for i in x:
    j=numpy.arcsin(i/R_4)
    Theta_1.append(j)
print("***************************************")
print("This is Theta_1")
print(Theta_1)

# Calculation of Theta_2
for i in Theta_1:
    k=numpy.arcsin((n_air/n_acr)*numpy.sin(i))
    Theta_2.append(k)
    
# Calculation of l_1     
for i in Theta_2:
    m=R_4*math.cos(i)-math.sqrt((R_4*math.cos(i))**2-(R_4**2-R_3**2))
    l_1.append(m)
print("****************************************")
print("This is Theta_2: ")
print(Theta_2)
print("****************************************")
print("This is l_1: ")
print(l_1)

# Calculation of Phi_1
for i in l_1:
    n=numpy.arccos((R_3**2+R_4**2-i**2)/(2*R_3*R_4))
    Phi_1.append(n)

print("***************************************")
print("This is Phi_1")
print(Phi_1)

# Calculation of Phi_2

for i in Phi_1:
    o=R_4*math.sqrt(2*(1-math.cos(i)))
    d_1.append(o)
print("***************************************")
print("This is d_1")
print(d_1)

#--------------------------------------------------------------
d=[d_1,l_1]
export_data=zip_longest(*d, fillvalue='')
with open("Theta3.xlsx", "w", encoding="ISO-8859-1", newline='') as First_layer:
    wr=csv.writer(First_layer)
    wr.writerow(("d_1","l_1"))
    wr.writerows(export_data)
First_layer.close()
#-------------------------------------------------------------

# Now we read those data from the new spreadsheet
New_data_theta3=pd.read_csv("Theta3.xlsx")

for i in numpy.arange(0, len(New_data_theta3["d_1"]), 1):
          p=numpy.arccos(((R_4-R_3)**2+ (New_data_theta3.at[i,'l_1'])**2-(New_data_theta3.at[i,'d_1'])**2)/(2*(New_data_theta3.at[i,'l_1'])*(R_4-R_3)))
          Theta_3.append(p)
print("***************************************")
print(len(Theta_1))
print(len(Theta_2))
print(len(Theta_3))
print("***************************************")

print(Theta_3)
```

```python
import pandas as pd
import csv
from itertools import zip_longest
Delta_12=[]
Delta_34=[]
Delta_56=[]
first_element=[]
second_element=[]
third_element=[]

Final_result=[]
Refraction=[]

for i in Theta_1:
    for j in Theta_2:
        if Theta_1.index(i)==Theta_2.index(j):
            p=numpy.sin(i-j)
            Delta_12.append(p)

for i in Theta_3:
    for j in Theta_4:
        if Theta_3.index(i)==Theta_4.index(j):
            q=numpy.sin(i-j)
            Delta_34.append(q)
            
for i in Theta_5:
    for j in Theta_6:
        if Theta_5.index(i)==Theta_6.index(j):
            r=numpy.sin(i-j)
            Delta_56.append(r)
            
for num1 , num2 in zip(l_1,Delta_12):
    first_element.append(num1*num2)
    
    
    
for num1 , num2 in zip(l_2,Delta_34):
    second_element.append(num1*num2)
    
    
    
for num1 , num2 in zip(l_3,Delta_56):
    third_element.append(num1*num2)

    

print("This is x_final:  ")
print('******************************************************************************')
print(third_element)
print('******************************************************************************')

zip_file=zip(x, first_element, second_element, third_element)
for num1 , num2 , num3 , num4 in zip_file:
    Final_result.append(num1-num2+num3-num4)
print("This is the final result:   ")
print("******************************************************************************")
print(Final_result)
print("******************************************************************************")


zip_diff=zip(x, Final_result)
for num1, num2 in zip_diff:
    Refraction.append(numpy.abs(num1-num2))
print("This shows the refraction of the photon inside the layers: ")
print("***************************************************************************************************")
print(Refraction)
print("***************************************************************************************************")

d=[x, Final_result]
export_data=zip_longest(*d, fillvalue='')
with open("acry_two_cent.csv", "w", encoding="ISO-8859-1", newline='') as acr_two:
    wr=csv.writer(acr_two)
    wr.writerow(("Position_2","Final_result"))
    wr.writerows(export_data)
acr_two.close()

        

```
